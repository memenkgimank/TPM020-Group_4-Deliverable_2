# Product Data Variance Check (Reliability of The Products)
Categorized into two subset of ratings:
Good (4 to 5 star);
Bad (1 to 3 star)


```python
# Fill missing star percentages with 0 (some products have NaN)
product_data = product_data.fillna(0)

# Calculate good vs bad
product_data["Good"] = product_data["Percent 4 star"] + product_data["Percent 5 star"]
product_data["Bad"] = product_data["Percent 1 star"] + product_data["Percent 2 star"] + product_data["Percent 3 star"]
```

```python
product_data["Variance_score"] = abs(product_data["Good"] - product_data["Bad"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate Good (4 & 5 stars) and Bad (1–3 stars)
product_data["Good"] = product_data["Percent 4 star"] + product_data["Percent 5 star"]
product_data["Bad"] = (
    product_data["Percent 1 star"] +
    product_data["Percent 2 star"] +
    product_data["Percent 3 star"]
)

# Variance score = absolute difference between good and bad
product_data["Variance_score"] = abs(product_data["Good"] - product_data["Bad"])

# Plot distribution
# Plot histogram with counts
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(product_data["Variance_score"], bins=30, edgecolor="black")

# Add counts on top of bars
for count, bin_left, bin_right in zip(counts, bins[:-1], bins[1:]):
    if count > 0:  # only label bars that exist
        plt.text(
            (bin_left + bin_right) / 2,  # bar center
            count,                       # height
            str(int(count)),             # label
            ha="center", va="bottom", fontsize=8
        )

plt.hist(product_data["Variance_score"], bins=30, edgecolor="black")
plt.title("Distribution of Rating Variance Across Products")
plt.xlabel("Variance Score (Good (4 to 5 star)% - Bad (1 to 3 star)%)")
plt.ylabel("Number of Products")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Distribution of Rating Variance without Subset of Rating

```python
# Collect the rating percentage columns
rating_cols = ["Percent 1 star", "Percent 2 star", "Percent 3 star",
               "Percent 4 star", "Percent 5 star"]

# Calculate variance across all 5 star percentages
product_data["Variance_score"] = product_data[rating_cols].var(axis=1)

# Plot histogram with counts
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(product_data["Variance_score"], bins=30, edgecolor="black")

# Add counts on top of bars
for count, bin_left, bin_right in zip(counts, bins[:-1], bins[1:]):
    if count > 0:
        plt.text((bin_left + bin_right)/2, count, str(int(count)),
                 ha="center", va="bottom", fontsize=8)

plt.title("Distribution of Rating Variance")
plt.xlabel("Variance Score (spread of rating distribution)")
plt.ylabel("Number of Products")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
